```python
#coding:utf-8
import numpy as np
from matplotlib.pylab import *
from scipy import fftpack
%matplotlib inline
```

```python
N = 512
```

```python
hamming_window = np.hamming(N)
hanning_window = np.hanning(N)
bartlett_window = np.bartlett(N)
blackman_window = np.blackman(N)
kaiser_window = np.kaiser(N, 5)
```

```python
subplot(231)
plot(hamming_window)
title("Hamming Window")
axis((0, N, 0, 1))

subplot(232)
plot(hanning_window)
title("Hanning Window")
axis((0, N, 0, 1))

subplot(233)
plot(bartlett_window)
title("Bartlett Window")
axis((0, N, 0, 1))

subplot(234)
plot(blackman_window)
title("Blackman Window")
axis((0, N, 0, 1))

subplot(235)
plot(kaiser_window)
title("Kaiser Window")
axis((0, N, 0, 1))

show()
```

### 窓関数の適用実験

```python
# 設定
N = 512
dt = 0.01
f = 10
t = np.linspace(1, N, N) * dt - dt
y = np.sin(np.pi * f * t)
```

```python
# windowedでハミング窓を乗算
original = y
ham_windowed = hamming_window * y
han_windowed = hanning_window * y
bar_windowed = bartlett_window * y
```

```python
# 波形描画
subplot(1,4,1)
plot(t,original)
xlabel("time")
ylabel("amplitude")

subplot(1,4,2)
plot(t,ham_windowed)
xlabel("time")
ylabel("amplitude")

subplot(1,4,3)
plot(t,han_windowed)
xlabel("time")
ylabel("amplitude")

subplot(1,4,4)
plot(t,bar_windowed)
xlabel("time")
ylabel("amplitude")
```

```python
# 周波数領域と位相領域を見る
originalDFT = fftpack.fft(original)
ham_windowedDFT = fftpack.fft(ham_windowed)
han_windowedDFT = fftpack.fft(han_windowed)
bar_windowedDFT = fftpack.fft(bar_windowed)
```
